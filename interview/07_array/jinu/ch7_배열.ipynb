### 7번. 두 수의 합
덧셈하여 타겟을 만들 수 있는 두 숫자 인덱스를 리턴하라.  
input : nums = [2, 7, 11, 15], target = 9.  
output : [0,1]

```python
# sol1. 내 풀이

nums = [2, 7, 11, 15]
target = 9

def twosum(nums, target):
    for i1, n1 in enumerate(nums):
        for i2, n2 in enumerate(nums):
            if n1 + n2 == target:
                return [i1, i2]
```

```python
result = twosum(nums, target)
print(result)
```

```python
# sol5. 책 풀이(투포인터)
# 입력값 nums의 정렬이 필요 -> 정렬을 이용하면 인덱스의 순서가 뒤섞이기 때문에 불가능

nums = [2, 7, 11, 15]
target = 9


def twosum(nums, target):
    left, right = 0, len(nums)-1

    while not left == right:
        if nums[left] + nums[right] < target:
            left += 1
        elif nums[left] + nums[right] > target:
            right -= 1
        else:
            return [left, right]
```

### 8번. 빗물 트래핑
높이를 입력받아 비 온 후 얼마나 많은 물이 쌓일 수 있는지 계산하기.  
input = [0,1,0,2,1,0,1,3,2,1,2,1].  
output = 6

```python
# sol1. 투 포인터 이동

height = [0,1,0,2,1,0,1,3,2,1,2,1]

if not height:
    print('None')

volume = 0
left, right = 0, len(height) -1
left_max, right_max = height[left], height[right]

while left < right:
    left_max, right_max = max(height[left], left_max), max(height[right], right_max)

    if left_max <= right_max:
        volume += left_max - height[left]
        left += 1
    else:
        volume += right_max - height[right]
        right -= 1
print(volume)
```

```python
# sol2. 스택 쌓기

stack = []
volume = 0

for i in range(len(height)):
    while stack and height[i] > height[stack[-1]]: # height[stack[-1]] : 마지막 height 높이
        top = stack.pop() # 바로 전 height

        if not len(stack):
            break

        distance = i - stack[-1] -1
        waters = min(height[i], height[stack[-1]]) - height[top]

        volume += distance * waters

    stack.append(i)

print(volume)
```

### 10번. 배열 파티션1
n개의 페어를 이용한 min(a,b)의 합으로 만들 수 있는 가장 큰 수를 출력하라.  
input = [1,4,3,2]  
output = 4

```python
# sol1. 오름차순 풀이

nums = [1,4,3,2] 

sum = 0
pair = []
nums.sort() # [1,2,3,4]

for n in nums:
    pair.append(n) #1. [1,2] #2. [3,4]
    if len(pair) == 2:
        sum += min(pair) #1. sum = 1 #2. sum = 4
        pair = [] 

print(sum)
```

```python
# sol2. 짝수 번째 값 계산

nums = [1,4,3,2]

sum = 0
nums.sort()

for i,n in enumerate(nums):
    if i%2 == 0:
        sum += n

print(sum)
```

```python
# sol1

prices = [7,1,5,4,6,4]

max_price = 0

for i, price in enumerate(prices):
    for j in range(i, len(prices)):
        max_price = max(prices[j] - price, max_price)        

print(max_price)
```
