### 1번. 유효한 팰린드롬

```python
a = 'A man, a plan, a canal: Panama'
b = 'race a car'
```

```python
# sol1

strs = []
for char in a:
    if char.isalnum():
        strs.append(char.lower())

```

```python
while len(strs) > 1:
    if strs.pop(0) == strs.pop():
        print('True')

    else:
        print('False')
```

```python
def isPalindrome(s):
    strs = []
    for char in s:
        if char.isalnum():
            strs.append(char.lower()) 
    while len(strs)>1:
        if strs.pop(0) != strs.pop():
            return False

    return True
    
```

```python
isPalindrome(a)
```

```python
isPalindrome(b)
```

```python
# sol2. deque이용
import collections

def isPalindrome2(s):
    strs = collections.deque()

    for char in s:
        if char.isalnum():
            strs.append(char.lower())

    while len(strs)>1:
        if strs.popleft() != strs.pop():
            return False
            
    return True

```

```python
isPalindrome2(a)
```

```python
isPalindrome2(b)
```

```python
# sol3. 슬라이싱 이용
import re

def isPalindrome3(s):
    s = s.lower()
    s = re.sub('[^a-z0-9]', '', s)

    return s == s[::-1]
```

```python
isPalindrome3(a)
```

```python
isPalindrome3(b)
```

### 3번. 로그파일 재정렬

```python
logs = ["dig1 8 1 5 1","let1 art can","dig2 3 6","let2 own kit dig","let3 art zero"]
```

```python
def reorderLogFiles(logs):
    letters, digits = [], []
    for log in logs:
        if log.split()[1].isdigit():
            digits.append(log)
        else:
            letters.append(log)
    letters.sort(key=lambda x: (x.split()[1:], x.split()[0]))
    return letters + digits
    
```

```python
reorderLogFiles(logs)
```

### 4번. 가장 흔한 단어

```python
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit." 
banned = ["hit"]
```

```python
words = re.sub('[^\w]', ' ', paragraph).lower().split()
# words = words.lower().split()
words
```

```python
counts = collections.Counter(result)
counts
```

```python
counts.most_common(1)[0][0]
```

```python
def mostCommonWord(paragraph, banned):
    words = re.sub('[^\w]', ' ', paragraph).lower().split()
    result = [word for word in words if word not in banned]
    counts = collections.Counter(result)
    return counts.most_common(1)[0][0]
```

```python
mostCommonWord(paragraph, banned)
```

### 5번. 그룹 애너그램

```python
strs = ["eat","tea","tan","ate","nat","bat"]
```

```python
def groupAnagrams(strs):
    anagrams = collections.defaultdict(list) # keyerror 방지
    
    for word in strs:
        anagrams[''.join(sorted(word))].append(word)
    return list(anagrams.values())
```

### 6번. 가장 긴 팰린드롬 부분 문자열

```python
def longestPalindrome(s):
    def expand(left, right): # 포인터를 옮기는 함수
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left+1:right]

    if len(s)<2 or s==s[::-1]:
        return s

    result = ''
    for i in range(len(s)-1):
        result = max(result, expand(i,i+1), expand(i,i+2), key=len) # 짝수 팰린드롬, 홀수 팰린드롬
        
    return result
```

```python
def expand(sen, left, right): # 포인터를 옮기는 함수
    while left >= 0 and right < len(sen) and sen[left] == sen[right]:
        left -= 1
        right += 1
    return sen[left+1:right]
```
