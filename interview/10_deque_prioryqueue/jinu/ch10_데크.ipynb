### 26번. 원형 데크 디자인


```python
class ListNode:
    def __init__(self, val, rightnode = None, leftnode = None):
        self.val = val
        self.right = rightnode
        self.left = leftnode
```

```python
class MyCircularDeque():
    def __init__(self, k):
        self.head = ListNode(None)
        self.tail = ListNode(None)
        self.k = k
        self.len = 0
        self.head.right = self.tail 
        self.tail.left = self.head

    # 이중 연결 리스트에 신규 노드 삽입
    def _add(self, node, new):
        n = node.right
        node.right = new
        new.left, new.right = node, new
        n.left = new

    def _del(self, node):
        n = node.right.right
        node.right = n
        n.left = node
    
    def insertFront(self, value:int):
        if self.len == self.k:
            return False
        self.len += 1
        self._add(self.head, ListNode(value))
        return True
    
    def deleteFront(self):
        if self.len == 0:
            return False
        self.len -= 1
        self._del(self.head)
        return True
    
    def deleteLast(self):
        if self.len == 0:
            return False
        self.len -= 1
        self._del(self.tail.left.left)
        return True

    def getFront(self):
        return self.head.right.val if self.len else -1

    def getRear(self):
        return self.tail.left.val if self.len else -1

    def isEmpty(self):
        return self.len == 0

    def isFull(self):
        return self.len == self.k
```

### 27번. k개 정렬 리스트 병합
k개의 정렬된 리스트를 1개의 정렬된 리스트로 병합하라.  
- input : [
    1 -> 4 -> 5,
    1 -> 3 -> 4,
    2 -> 6
]  

- output : 1 -> 1 -> 2 -> 3 -> 4 -> 4 -> 5 -> 6


참고 : https://byumm315.tistory.com/entry/%EB%A6%AC%ED%8A%B8%EC%BD%94%EB%93%9C-23-Merge-k-Sorted-Lists-1


```python
import heapq

def mergeKLists(lists):
    root = result = ListNode(None)
    heap = []

    # 각 연결 리스트의 루트를 힙에 저장
    for i in range(len(lists)):
        if lists[i]:
            # heap에 저장
            heapq.heappush(heap, (lists[i].val, i, lists[i])) # ex) lists[0] = [1 -> 4 -> 5] -> (1, 0, 1 -> 4 -> 5), heap = [(1,0,1->4->5), (1,1,1->3->4), (2,2,2->6)]

    # 힙 추출 이후 다음 노드는 다시 저장
    while heap: # 힙이 빌 때까지

        # heappop -> 가장 작은 원소부터 오름차순으로 heap의 원소를 추출후 heap에서 제거
        node = heapq.heappop(heap) # ex) heap = [(1,0,1->4->5), (1,1,1->3->4), (2,2,2->6)]일 때, (1,0,1->4->5) 추출. node = (1,0,1->4->5)
        idx = node[1] # ex) idx = 0
        result.next = node[2] # ex) result.next = (1->4->5)

        result = result.next
        if result.next:
            heapq.heappush(heap, (result.next.val, idx, result.next)) # ex) heap = [(4,0,4->5), (1,1,1->3->4), (2,2,2->6)]
        
    return root.next

```
